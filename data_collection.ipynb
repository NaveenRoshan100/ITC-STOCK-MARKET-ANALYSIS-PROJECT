```python
import pandas as p
import os
```

```python
API_KEY = os.environ['APLHA']
symbol = "ITC.BSE"
url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={symbol}&apikey={API_KEY}&datatype=csv&outputsize=full"
data=p.read_csv(url)
```

```python
data.to_csv('ITC-STOCK-DATA.csv',index=False)
```
